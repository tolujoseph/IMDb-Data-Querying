```python
#Importing libraries
import numpy as np 
import pandas as pd 
import sqlite3 as sql

import os
for dirname, _, filenames in os.walk(r'C:/Users/IMDb'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#Importing data-set
db = r'C:/Users/IMDb/movies.sqlite'
connection = sql.connect(db) 

print("SQL database connected")
table = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'", connection)
print(table)
```

```python
#Can I get all data about movies?
query = """
    SELECT *
    FROM movies;
"""
movies_all = pd.read_sql(query, connection)
movies_all
```

```python
#How do I get all data about directors?
query = """
    SELECT *
    FROM directors;
"""
directors_all = pd.read_sql(query, connection)
directors_all
```

```python
#Check how many movies are present in iMDB.
query = """
    SELECT count (*)
    FROM movies;
"""
movies_all = pd.read_sql(query, connection)
movies_all
```

```python
#Find these 3 directors: James Cameron ; Luc Besson ; John Woo
query = """
    SELECT *
    FROM directors
    Where name =='James Cameron' or name=='Luc Besson' or name=='John Woo' ;
"""
directors_selected = pd.read_sql(query, connection)
directors_selected
```

```python
#Find all directors with name starting with Steven
query = """
    SELECT *
    FROM directors
    Where name like 'Steven%';
"""
directors_selected = pd.read_sql(query, connection)
directors_selected
```

```python
#Count female directors.
query = """
    SELECT count(*)
    FROM directors
    Where gender=='1';
"""
directors_female = pd.read_sql(query, connection)
directors_female
```

```python
#Find the name of the 10th first women directors?
query = """
    SELECT name
    FROM directors
    Where gender=='1'
    limit 10;
"""
directors_female = pd.read_sql(query, connection)
directors_female
```

```python
#What are the 3 most popular movies?
query = """
    SELECT *
    FROM movies
    order by popularity desc
    limit 3;
"""
movies_all = pd.read_sql(query, connection)
movies_all
```

```python
#What are the 3 most bankable movies?
query = """
    SELECT *
    FROM movies
    order by revenue desc
    limit 3;
"""
movies_all = pd.read_sql(query, connection)
movies_all
```

```python
#What is the most awarded average vote since the January 1st, 2000?
query = """
    SELECT vote_average
    FROM movies
    where release_date > '2000-01-01'
    ORDER by vote_average DESC
    limit 1;
"""
movies_all = pd.read_sql(query, connection)
movies_all
```

```python
#Which movie(s) were directed by Brenda Chapman?
query = """
    SELECT m.original_title
    FROM movies m
    Join directors d
    ON m.director_id=d.id
    where name = 'Brenda Chapman'
"""
movies_all = pd.read_sql(query, connection)
movies_all
```

```python
#Whose director made the most movies?
query = """
    SELECT name,count(name)
    FROM movies m
    Join directors d
    ON m.director_id=d.id
    group by director_id
    order by count(name) DESC;
"""
movies_all = pd.read_sql(query, connection)
movies_all
```

```python
#Whose director is the most bankable?
query = """
    SELECT name,revenue
    FROM movies m
    Join directors d
    ON m.director_id=d.id
    group by director_id
    order by revenue DESC;
"""
movies_all = pd.read_sql(query, connection)
movies_all
```
